# 函数定义

函数是描述两个变量之间关系的一种数学方式
- 具体来说 函数会把一个输入(**自变量**)映射到一个输出(**因变量**)
- 一个机器,它接收一个数字作为输入,经过处理后输出另一个数字.
- 如果你把输入叫做$x$,输出叫做$f(x)$,那么这个机器的工作方式就可以用函数$f(x)$来表示
    - $f(x)=2x+3$
    - 输入$x$会被乘以2,再加上3,输出就是$f(x)$

# 函数图像
函数图像就是在平面直角坐标系上表示一个函数的方式.

每一个函数值$f(x)$对应着$x$的一个点
1. 横纵通常表示自变量$x$
2. 纵轴通常表示因变量$f(x)$

```python
def f(x):
    return x ** 2

x_values = np.linspace(-10,10,400)
y_values = f(x_values)

plt.plot(x_values,y_values,label='f(x) = x^2')
plt.show()
```
