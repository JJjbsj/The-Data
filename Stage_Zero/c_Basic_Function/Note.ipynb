```python
# import 
import numpy as np
import matplotlib.pyplot as plt
```

# 函数定义

函数是描述两个变量之间关系的一种数学方式
- 具体来说 函数会把一个输入(**自变量**)映射到一个输出(**因变量**)
- 一个机器,它接收一个数字作为输入,经过处理后输出另一个数字.
- 如果你把输入叫做$x$,输出叫做$f(x)$,那么这个机器的工作方式就可以用函数$f(x)$来表示
    - $f(x)=2x+3$
    - 输入$x$会被乘以2,再加上3,输出就是$f(x)$

简单来说就是想python一样做一个函数出来

更直观的理解就是你有一个工具箱,里面有非常多已经做好的工具了,这个就是函数,你只是需要使用而已

但是你需要先输入它才会输出

# 函数图像
函数图像就是在平面直角坐标系上表示一个函数的方式.

**每一个函数值$f(x)$对应着$x$的一个点**
1. 横纵通常表示自变量$x$     = 输入
2. 纵轴通常表示因变量$f(x)$  = 输出

简单来说就是,把函数的输入和输出绘制成为一个图像

```python
def f(x):
    return x ** 2

x_values = np.linspace(-10,10,400)
y_values = f(x_values)

plt.plot(x_values,y_values,label='f(x) = x^2')
plt.grid()
plt.title("Graph of f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

# 线性函数(斜率/截距)
$$f(x) = ms + b $$
- $m$是斜率,表示每单位$x$增加时,函数值$f(x)$增加的量
    - 斜率越大,函数图像越陡峭
- $b$是截距,表示函数图像与纵轴的交点(当$x=0$时,$f(x)=b$)

例子

$f(x) = 2x +3$
- 斜率$m$ = 2,意味着每增加1个单位的$x$,函数值$f(x)$会增加2个单位
- 解决$b$ = 3,意味着当$x = 0$时,$f(x)=3$

这个函数的图像时一条直线,斜率为2,纵轴交点在3

> 简单来说,线性函数的含义就是变量与变量之间的线性关系
> 
> x增加,y会是怎么样变化的.

```python
def f(m,x,b):
    return  m * x + b

x = np.linspace(0,10,400)

y = f(10,x,0)

plt.plot(x, y, label=f"f(x) = mx + b", color='blue')

# 添加图像标题和标签
plt.title("Graph of f(x) = mx + b")
plt.xlabel("x")
plt.ylabel("f(x)")

# 显示网格
plt.grid(True)

# 显示图像
plt.legend()
plt.show()
```

# 二次函数(顶点形式,标准形式)
$$ f(x) = ax^2 + bx + c$$
- a,b,c是常数
- a控制图像的开口方向
    - a > 0 开口向上
    - a < 0 开口向下

## 顶点形式
二次函数还可以用顶点形式来表示: $ f(x) = a(x-h)^2 + k$
- (h,k)是顶点,函数图像的最高点和最低点

例子:

如果$f(x)=2(x-1)^2 + 3$
- 顶点是(1,3)
- 开口向上,因为$a = 2 > 0$

```python
def f(x):
    return 2*(x-1)**2 + 3

x_values = np.linspace(-3, 5, 400)
y_values = f(x_values)

plt.plot(x_values, y_values, label=r'$f(x) = 2(x - 1)^2 + 3$', color='blue')

plt.scatter(1, 3, color='red', label="Vertex (1, 3)")

plt.title("Graph of f(x) = 2(x - 1)^2 + 3")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.grid(True)

plt.legend()
plt.show()
```

# 指数函数(增长/衰减)
$$ f(x)=a \cdot b^x$$

1. a是常数,决定图像的纵轴截距
2. b是底数,决定函数的增长或者衰减速度
    - 如果b > 1,函数是增长的
    - 如果b < 1,函数是衰减的

例子
1. 假设$f(x)=2 \cdot 3^x$
    - 当x增加时,$f(x)$会快速增长
1. 如果$f(x)=5 \cdot (0.5)^x$
    - 当x增加时,$f(x)$会逐渐减少

```python
def growth_function(x):
    return 2 * 3**x

def decay_function(x):
    return 5 * (0.5)**x

x_values = np.linspace(-2, 5, 400)

y_growth = growth_function(x_values)
y_decay = decay_function(x_values)

plt.plot(x_values, y_growth, label=r'$f(x) = 2 \cdot 3^x$', color='blue')

plt.plot(x_values, y_decay, label=r'$f(x) = 5 \cdot (0.5)^x$', color='red')

plt.title("Exponential Growth and Decay")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.grid(True)

plt.legend()
plt.show()
```

# 对数函数(性质,对数律)
对数函数是指数函数的逆运算
$$
    f(x) = log_b(x)
$$
这里b是对数的底数,通常我们讨论两种常见的对数
1. 常用对数,底数$b = 10$ 记作$log(x)$
2. 自然对数,底数$b = e$(e是自然常数,约等于2.718),记作$ln(x)$

对数的性质
1. 对数的基本运算
    - $ log_b(xy) = log_b(x) + log_b(y)$
    - $ log_b(\frac{x}{y}) = log_b(x) - log_b(y)$
    - $ log_b(x^n) = n \cdot log_b(x)$
2. 对数函数的图像
    - 对数函数的图像是一个曲线
        1. 当x趋近于0时,函数趋向负无穷
        2. 当x增加时,函数值缓慢增长

例子
1. 如果$f(x)=log_2(x)$
    - 当$x = 2$时,$f(x)=1$
    - 当$x = 4$时,$f(x)=2$

```python
def log_function(x, base=2):
    return np.log(x) / np.log(base)  

x_values = np.linspace(0.1, 10, 400)

y_values_base2 = log_function(x_values, base=2)
y_values_base10 = log_function(x_values, base=10)

plt.plot(x_values, y_values_base2, label=r'$f(x) = \log_2(x)$', color='blue')
plt.plot(x_values, y_values_base10, label=r'$f(x) = \log_{10}(x)$', color='red')

plt.title("Graph of Logarithmic Functions")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.grid(True)

plt.legend()
plt.show()
```

# 函数总结
1. **函数定义** 是输入和输出之间的关系
1. **线性函数** 是斜率和截距的函数
1. **二次函数** 描述抛物线的形状，可以用标准形式和顶点形式表示
1. **指数函数** 表示快速增长或衰减
1. **对数函数** 是指数函数的逆运算，包含一些基本的对数性质
