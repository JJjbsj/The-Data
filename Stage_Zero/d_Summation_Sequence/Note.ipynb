### 1. **求和符号 Σ (Sigma)**

求和符号 Σ 是一个常用的数学符号，用来表示一系列数值的总和。这个符号的来源是希腊字母 "Sigma"（Σ），它的意思就是“加总”。

#### **一般形式**：

$$
\sum_{i=a}^{b} f(i)
$$

* **$i$** 是求和的变量
* **$a$** 是求和的起始值 
* **$b$** 是求和的终止值
* **$f(i)$** 是需要求和的函数或数列

例如：
$$
\sum_{i=1}^{5} i = 1 + 2 + 3 + 4 + 5 = 15
$$
这里，**$i$** 从 1 加到 5，总和是 15。

### 2. **算术级数 (Arithmetic Sequence)**

算术级数是指每一项和前一项的差是常数的数列。常数差叫做**公差**（d）。

#### **一般形式**：

* 第n项：($ a_n = a_1 + (n-1) \cdot d$ )

  * 其中 ($a_1$) 是首项，d 是公差，n 是项数。

#### **算术级数求和公式**：

如果我们要求从 ($a_1$) 到 ($a_n$) 这些项的和，可以使用下面的公式：
$$
S_n = \frac{n}{2} \cdot (a_1 + a_n)
$$
或者
$$
S_n = \frac{n}{2} \cdot (2a_1 + (n-1) \cdot d)
$$
其中：

* ($S_n$) 是前n项的和
* ($a_1$) 是首项
* ($a_n$) 是第n项
* ($d$) 是公差
* ($n$) 是项数

**例子**：
给定算术级数：2, 5, 8, 11, 14，求前 5 项的和。

* 首项 ($a_1 = 2$)
* 公差 ($d = 3$)
* 项数 ($n = 5$)

根据公式：
$$
S_5 = \frac{5}{2} \cdot (2 + 14) = \frac{5}{2} \cdot 16 = 40
$$
所以，前 5 项的和是 40。

### 3. **几何级数 (Geometric Sequence)**

几何级数是指每一项和前一项的比是常数的数列。这个常数比叫做**公比**（r）。

#### **一般形式**：

* 第n项：( $a_n = a_1 \cdot r^{n-1}$ )

  * 其中 ($a_1$) 是首项，r 是公比，n 是项数。

#### **几何级数求和公式**：

对于一个有n项的几何级数，从 ($a_1$) 开始，公比为 r 的情况下，总和是：
$$
  S_n = a_1 \cdot \frac{1-r^n}{1-r} (当 r ≠ 1)
$$
如果 (r = 1)，所有项都是相同的，求和公式就变成了：
$$
S_n = n \cdot a_1
$$

**例子**：
给定几何级数：2, 6, 18, 54，公比 (r = 3)，求前 4 项的和。

* 首项 ($a_1 = 2$)
* 公比 ($r = 3$)
* 项数 ($n = 4$)

根据公式：
$$
S_4 = 2 \cdot \frac{1 - 3^4}{1 - 3} = 2 \cdot \frac{1 - 81}{-2} = 2 \cdot \frac{-80}{-2} = 80
$$
所以，前 4 项的和是 80。

级数不仅是把数加起来，而且这串数通常有规律

算术级数：每一项比前一项多一个固定数，比如 1、3、5、7、9……

几何级数：每一项是前一项乘一个固定数，比如 2、4、8、16……

```python
# -------------------------
# 算术级数示例（固定增长）
# -------------------------
a1 = 1000  # 第一年存入 1000
d = 1000   # 每年增加 1000
n = 5      # 存5年

print('第一年存入1000')
print('每年增加1000')
print('一共5年')
print("-----")

# 用公式求和
# S_n = n/2 * (a1 + a_n)
# 先算第 n 年的金额
an = a1 + (n-1)*d
S_arithmetic = n/2 * (a1 + an)

print("算术级数示例：")
print(f"第 {n} 年的存入金额: {an}")
print(f"{n} 年总存入金额: {S_arithmetic}")
print("-----")

# -------------------------
# 几何级数示例（复利增长）
# -------------------------
a1 = 1000  # 第一年存入 1000
r = 1.1    # 每年按10%复利增长
n = 5      # 存5年

# 第 n 年金额
an = a1 * r**(n-1)
# 前 n 年总和
S_geometric = a1 * (1 - r**n) / (1 - r)  # r != 1

print("几何级数示例：")
print(f"第 {n} 年的存入金额: {an:.2f}")
print(f"{n} 年总存入金额: {S_geometric:.2f}")
```


### 4. **有限级数求和公式**

有限级数的求和是指有限个数列项的总和。

对于算术级数和几何级数，我们已经提供了常用的求和公式。

对于更一般的级数，如果是其他形式的数列，可以根据具体形式采取相应的方法。

例如，求和公式的推导和应用一般都和数列的结构紧密相关，比如常见的特殊数列（如斐波那契数列等）会有专门的求和技巧。

1. **Σ符号**用于表示数列的求和，具体范围和表达式通过下标确定。

3. **几何级数**的关键是公比，求和公式在公比不等于1时，通过首项和公比的关系计算总和。
