### 一、变量与常数（Variables & Constants）

**常数（Constant）**
固定的数字，不会随着情况的发展而改变
→ 静态量

**变量（Variable）**
不确定的数字，或会随条件变化的数字
→ 动态量

**用途说明**

* 用来描述现实世界中的数量关系
* 常用于公式、模型、规则说明
* 不一定立即参与计算，也可以只是概念描述

> 这一层解决的问题是：**哪些东西会变，哪些不会变**

---









### 二、代数表达式（Algebraic Expression）

**基础构成元素**

* 数字：常数
* 字母：变量
* 运算符：`+ - × ÷`（底层计算逻辑）
* 括号：表示运算层级
* 运算规则：决定计算顺序与结果

**关键区分**

* 代数表达式 ≠ 等式

**代数表达式**

* 描述一个公式的结构
* 例子: $\frac{p_t - p_{t-1}}{p_t-1}$

**等式**

* 判断两个值是否相等
* 例子：
  $
  3x + 7 = 13
  $

> 本质区别：
> 一个在**描述结果**，一个在**判断关系**

---

### 三、表达式的化简（Simplification）

**化简的目标**

* 不改变原有含义
* 用更清晰、更短的方式表达

**核心只有两件事**

1. 合并相同结构的项
2. 展开括号中的重复结构

---

**合并同类项**
$
5x + 4x = 9x
$

解释：

* 描述的是同一种东西（x）
* 只是数量不同
* 可以合并计算

---

**分配律（展开括号）**
$
2(x + 3) = 2x + 6
$

解释：

* 括号里的每一项都要乘以外面的数
* 用于“拆清结构”，不是增加复杂度

---

### 四、多项式（Polynomial）

**项（Term）**
$
\text{项} = 系数 \times x^k
$

* 系数：数字部分
* 指数（幂）：变量被乘了几次

例子：

* (2x)：一次项
* (x^2)：二次项

---

**多项式的定义**

* 多个“项”通过加减连接
* 属于代数表达式

分类：

* 单项式：一个项
* 二项式：两个项（如 (3x + 5)）
* 多项式：三个及以上

---

### 五、多项式运算

**加减法**
$
(3x + 2) + (5x - 1)
$

步骤：

* 去括号
* 合并同类项

结果：
$
8x + 1
$

---





**乘法**
$
(x + 2)(x + 3)
$

规则：

* 没有符号连接，默认是乘法
* 每一项都要互相相乘

展开：
$
x^2 + 5x + 6
$

---


**除法**
$
\frac{x^2 + 5x + 6}{x + 2}
$

核心流程（始终只有三步）：

1. 最高次项相除
2. 商 × 除式
3. 被除式 − 结果

重复，直到不能再除

**多项式除法的核心就是找到一种系统的分配方式**
