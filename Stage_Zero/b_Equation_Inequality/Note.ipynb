## 线性方程(Liear Equations)

$$
    ax + b = y
$$

**简单解释**

线性方程就是一个简单的预测x会对于y产生结果的一个简单工具
- a是一个敏感性工具,也可以称其为杠杆,他的作用可以是非常大的正向帮助,也可以是反向帮助
- x是唯一一个可以自己控制的变量
- b是一个非常重要的对比参数,用它来对比y的结果是好还是差,b是一个起点
- y是一个终点,最终的预测位置

它的本质是一个可预测的,但是这个是一个简单的无脑预测,因为它只是简单把所有东西都加起来,然后就预测未来的情况

但是现实生活中,这个是一个不可能的情况,现实生活中充满非线性的时刻和情况.

---

**数学表达与运作逻辑**

两个计算方式
1. 根据axb计算c是多少(顺向求值)
2. 根据abc计算x是多少(逆向求解)

**顺向求值的计算很简单,只是简单加/乘而已**
1. $a \cdot x + b = c$

**逆向求解的计算分为两步**
$$ ax + b = c $$
1. 剥离干扰项(加减法逆运算)
    - 首先把不带x的项(常数b),移动到=的另一边
        - 注意移动需要变换运算符,变换到对立面(+/-..)
    - $ ax = c - b $
2. 还原单位项(乘法逆运算)
    - 我们要把x前面的系数a拿掉(和第一步一样,放到等式右边)
    - 同样需要变换运算符(×/÷)
    - $ x = \frac{c-b}{a}$

$$ x / 3 + 2 = 5$$
1. 首先第一步就是看清结构: $\frac{x}{3} + 2 = 5$
2. 然后常数项移项: $\frac{x}{3} = 5 -2$
    - $ \frac{x}{3} = 3$
3. 然后再做一次常数项移项: $x = 3 \times 3$(注意转换运算符)
4. $x=9$

> 移项的最终目标就是让X单独在一边,过程可以不规定移动哪边

---

$$ x+5=2x−3$$
1. 把x项移到一边
    - $ 5 = 2x − x − 3  $
    - $5 = x − 3$
2. 把常数项移到一边
    - $ 5 + 3 = x \Rightarrow x = 8$

```python
def plot_linear(a,b):
    # 生成 x 的数据点: 从 -10 到 10 ,生成 100 个数据点
    x = np.linspace(0,10,100)

    # 线性方程 y = ax + b
    y = a * x + b

    # 可视化绘制
    plt.figure(figsize=(15,5))
    plt.plot(x,y,label=f"y = {a}x + {b}",color='blue',linewidth=2)

    # 绘制坐标轴
    plt.axhline(0,color='black',linewidth=0.5)
    plt.axvline(0,color='black',linewidth=0.5)

    # 标注截距 b(当 x=0 时)
    plt.scatter(0,b,color='red',zorder=5)
    plt.annotate(f"Intercept (0, {b})",(0,b),textcoords="offset points",xytext=(10,10))

    plt.title("Visualizing Linear Equation")
    plt.xlabel("Input (x)")
    plt.ylabel("Output (y)")
    plt.grid(True,linestyle='--')
    plt.legend()
    plt.show()

# 尝试改变这些参数运行
# a 代表斜率(陡峭程度) b 代表截距(上下平移)
plot_linear(a=1,b=1)
```

```python
def explain_linear(a_high=5,b_high=10,a_low=1,b_low=0,effort=10):
    # 使用增长对比的例子
    x = np.linspace(0, effort, 100)
    
    # A
    y_high = a_high * x + b_high
    
    # B
    y_low = a_low * x + b_low

    plt.figure(figsize=(10, 6))
    plt.plot(x, y_high, label=f"A: y = {a_high}x + {b_high} ", color='green')
    plt.plot(x, y_low, label=f"B: y = {a_low}x + {b_low}", color='blue')
    
    plt.title("What does Linear Equation express?")
    plt.xlabel("Effort (x)")
    plt.ylabel("Result (y)")
    plt.legend()
    plt.grid(True)
    plt.show()

explain_linear()
```

## 一元二次方程(Quadraic Equations)
$$
    ax^2 + bx + c = 0
$$
这里最关键的表达式$x^2$
-输入的微小变化,会产生不成比例的巨大输出

1. $x^2$(二次项): 这是方程的核心,因为有了平方,它让变化不再是平滑的直线,而是带弯曲的曲线.
    - 决定了方程的 **形状**
2. $a$(二次项系数): 方程的 **"胖瘦/开口"**
    - 如果$a$很大,曲线会变得很陡峭(像个窄窄的V)
    - 如果$a$是正数,开口向上(像个笑脸);负数开口向下(像个哭脸)
3. $b$(一次项系数): 方程的偏移感,它和$a$同时决定了这条曲线的中心点(对称轴)在左边还是右边
4. $c$(常数项): 方程的起点,
    - 当$x=0$时,你在纵坐标的哪个位置.

---

**计算方法** : 
$$
    x^2 - 5x + 6 = 0
$$
1. 因式分解法 **(逻辑拆解)**
    - 第一步 == 拆解结果(常数项+6) **乘法**
        - 我们需要找到两个数,相乘等于+6的
            - $1 \cdot 6$
            - $(-1) \cdot (-6)$
            - $2 \cdot 3$
            - $(-2) \cdot (-3)$ 
    - 第二步 == 匹配过程(一次项系数 -5) **加法**
        - 我要从上面的选项里面找到一组数,加起来等于-5
        - 本质来说,我们就是需要找到一组数**相乘**符合常数项,**相加**符合一次项系数
        - $(-2) + (-3)$
    - 第三步 == 填入括号并解出答案
        - 找到了-2,-3我们直接把它填入括号
        - $ (x - 2)(x - 3)=0$
        - 我们需要它的答案是$x=0$,那么x本身需要是多少呢?    
            - 其实将得到的答案直接乘于-1,就是最终的答案了
        - $(x-2)=0 \rightarrow x = 2$
        - $(x-3)=0 \rightarrow x = 3$

---

$$
    x^2 + 2x - 8 = 0
$$
2. 求根公式 **(暴力求解)**
    - 当因式分解看不出来时,我们可以直接代入公式:
    - $x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$
    - 第一步 == 填表(认清a,b,c)
        - $a$是$x^2$前面的数(没写就是1)
        - $b$是$x$前面的数(这里是2)
        - $c$是最后的常数(这里是-8)
    - 第二步 == 算核心零件Δ:算好$b^2 - 4ac$是多少
        - $b^2 = 2 \cdot 2 = 4$
        - $4ac$
            - 先计算: $a \cdot c = 1 \cdot -8 = -8$
            - 再乘于4: $4 \cdot (-8) = -32$
        - 最后做减法: $\Delta = 4 - (-32) = 4 + 32 = 36$(负负得正)
        - 观察 $\Delta$ 是大于,小于,等于0
            - $> 0$ == 两个解
            - $ = 0 $== 一个解
            - $< 0 $== 无解
    - 把零件装进模板 : 
        - $x = \frac{-2 \pm \sqrt{36}}{2 \times 1}$
        - 因为 $\sqrt{36}=6$,所以 $x=\frac{-2 \pm 6}{2}$
    - 拆解为两个结果:(如果b本身是负数,那么就负负得正计算结果)
        - 一个用加法: $\frac{-2 + 6}{2} = \frac{4}{2} = 2$
        - 一个用减法: $\frac{-2 - 6}{2} = \frac{-8}{2} = -4$

---
**为什么$a$决定了形态?**
- 如果$a=0$,它就是我们之前学习的线性方程(匀速,没有疲劳,永远增长)
- 如果$a\neq0$世界变复杂了,它表示,事情的发展不是永无止境的,总有一种力量$(ax^2)$在拉扯

```python
def visualize_quadratic():
    # 设置投入精力 x 的范围：从 0 到 10
    x = np.linspace(0, 10, 400)
    
    # 定义三个不同的场景 (y = ax^2 + bx + c)
    # 场景 A: a=-0.5 (疲劳), b=5 (高效率), c=-8 (初始成本) -> 判别式 > 0
    y_success = -0.5 * x**2 + 5 * x - 8
    
    # 场景 B: a=-0.5, b=4, c=-8 -> 判别式 = 0 (刚好碰到 0)
    y_marginal = -0.5 * x**2 + 4 * x - 8
    
    # 场景 C: a=-0.5, b=2, c=-8 -> 判别式 < 0 (永远在 0 以下)
    y_failure = -0.5 * x**2 + 2 * x - 8

    plt.figure(figsize=(12, 7))
    
    # 绘制曲线
    plt.plot(x, y_success, label='Scenario A: Δ > 0 (Profitable Zone)', color='green', lw=2)
    plt.plot(x, y_marginal, label='Scenario B: Δ = 0 (Breakeven point)', color='orange', lw=2, linestyle='--')
    plt.plot(x, y_failure, label='Scenario C: Δ < 0 (Always in deficit)', color='red', lw=2)
    
    # 绘制盈亏平衡线 (y=0)
    plt.axhline(0, color='black', linewidth=1.5, label='Breakeven Line (y=0)')
    
    # 装饰图表
    plt.title("Visualizing Life Outcomes with Quadratic Equations", fontsize=15)
    plt.xlabel("Effort/Investment (x)", fontsize=12)
    plt.ylabel("Net Gain/Outcome (y)", fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.ylim(-15, 10)
    plt.legend()

    plt.show()

visualize_quadratic()
```

```python
def plot_quadratic_interactive(a, b, c):
    # 1. 环境准备
    x = np.linspace(-15, 15, 500)
    y = a*x**2 + b*x + c
    delta = b**2 - 4*a*c
    
    # 2. 创建画布
    fig = go.Figure()

    # 3. 绘制二次函数曲线
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='二次函数曲线', line=dict(color='blue', width=3)))

    # 4. 绘制地平线 (y=0)
    fig.add_trace(go.Scatter(x=[-15, 15], y=[0, 0], mode='lines', name='地平线 (y=0)', line=dict(color='black', dash='dash')))

    # 5. 计算并标记交点 (Roots)
    if delta >= 0:
        x1 = (-b + np.sqrt(delta)) / (2*a)
        x2 = (-b - np.sqrt(delta)) / (2*a)
        fig.add_trace(go.Scatter(x=[x1, x2], y=[0, 0], mode='markers', name='交点(解)', marker=dict(color='red', size=12)))

    # 6. 图表美化
    fig.update_layout(
        title=f"一元二次方程交互分析 (Δ = {delta})",
        xaxis_title="变量 X",
        yaxis_title="结果 Y",
        template="plotly_white",
        hovermode="x unified"
    )
    
    fig.show()

# 运行尝试：看看这个参数下的“结局”
plot_quadratic_interactive(a=1, b=5, c=6)
```

## 方程组求解(solving a system of equations)

**什么是方程组**

> 同一组未知数,同时满足多个方程

$$
\begin{cases} 
x + y = 5 \\
x - y = 1
\end{cases} 
$$

意思是：

找一对$(x,y)$，既满足第一个式子，又满足第二个式子

---

**方法一 == 代入法**

> 一个方程里把某个未知数"表示出来"
>
> 然后"代进"另一个方程,把问题变为一个未知数

例子
$$
    \begin{cases}
    x + y = 7 \\
    x - 2y = 1
    \end{cases}
$$
**第一步** : 
- 先选择一个方程,把某个未知数单独解出来(随便一个) 
- $ x = 7 - y $

**第二步** : 
- 把$x = 7 - y$代入第二个方程
- $ (7 - y) - 2y = 1$
- 注意这里的括号展开规则,如果括号前有"-",就需要转变括号内的运算符,如果没有就不需要
    1. 展开括号: $7 - y - 2y =1$
    2. 合并同类项: $7 - 3y = 1$
    3. 计算未知数 : $7 - 1 = 6 \Rightarrow 3y = 6 \Rightarrow 6 / 3 = 2 \Rightarrow y = 2 $

**第三步** : 
- 求另一个未知数
- $ x = 7 - y \Rightarrow x = 7 - 2 = 5 \Rightarrow x = 5$
- 最终解 : $(x,y) = (5,2)$'

---

**方法二 == 消元**
$$
    \begin{cases}
    ax + by = c \\
    dx + ey = f
    \end{cases}
$$
> 核心思想 = 把一个多维的方程压缩成为一维的方程,同时保留原有的约束


1. 第一步: 看系数和找到可以相互抵消的
    - 看方程组相同未知变量的系数
    - 找到一对系数,更容易变成「相同但一正一负」?(因为可以通过加减的方式去抵消为0)
    - 通常会有不同的情况
        1. 已经是一正一负,大小相同(最理想的状态)
            - $ \begin{cases} x + y = 5 \\ x - y = 1 \end{cases} $
            - 这里$+y,-y$,相加后,$y$直接消失
        2. 符号相同,大小相同
            - $\begin{cases} x + y = 5 \\ x + y = 1 \end{cases}$
            - 这里都是$+y$,所以直接相减$y$就消失.
        3. 大小不同(乘法)
            - $\begin{cases} 2x+3y = 12 \\ x-y=1\end{cases}$
            - 想要消除$y$,但是$+3y,-1y$
            - **把其中一个方程整体乘以一个数**,让系数变成3和-3
                - $ x-y=1 \times 3 \Rightarrow 3x - 3y = 3$
                - $ +3y + (-3y) = 0$
        4. 大小不同(除法)
            - $\begin{cases} 4x + 2y = 10 \\ 2x - y = 3 \end{cases}$
            - 第一行整体除以2: (通过缩小的方式也是一个方法)
                - $\begin{cases} 2x + y = 10 \\ 2x - y =3 \end{cases}$
3. 就是求被抵消掉的未知变量(回代)
    - 这一步就和上面的最后一步一样.
    - 可以随便选择一个方程进行回代.

---

例子示范 : $\begin{cases}4x + 2y = 10\\3x-y = 5\end{cases}$
1. 确定了抵消的未知数$y$,因为$y$的系数容易抵消
2. 调整方程系数
    - 第二个方程$3x - y=5$,为了让-1变成-2(抵消第一个方程的+2),**所以整体乘以2**
    - $ 2 \cdot (3x - y) = 2 \cdot 5 \Rightarrow 6x - 2y = 10$(整条方程都要乘)
    - 方程变为了$\begin{cases} 4x + 2y =10 \\ 6x - 2y =10\end{cases}$
3. 使用加减法抵消未知数
    - 这个阶段需要灵活应变去使用加法或者减法去抵消未知变量
    - $ (4x + 2y)+(6x-2y)=10+10$(这一步两个方程的不一定是加法的,根据抵消未知变量的运算符作为确定)
    - 然后合并同类项 **(合并同类项不是单纯的加起来,而是看项本身带的运算符去合并的)**
    - $6x + 0 = 10 + 10$
    - $10x = 20 \Rightarrow 20/10x = 2 \Rightarrow x = 2$
4. 解一元方程
    - 回代求另一个未知变量(选择哪一个方程都没有关系的)
    - $4x + 2y= 10$
    - $ 4 \cdot 2 + 2y = 10 \Rightarrow 8 + 2y = 10 \Rightarrow 2y = 2 \Rightarrow y = 1$
5. 最终解 : $(x,y) = (2,1)$

```python
def plot_linear_system_plotly(eq1, eq2, x_range=(-10,10), points=400):
    a1,b1,c1 = eq1
    a2,b2,c2 = eq2

    # 判别解类型
    det = a1*b2 - a2*b1
    if det == 0:
        print("方程组无唯一解，可能无解或无数解")
        return

    # 唯一解
    x_sol = (c1*b2 - c2*b1)/det
    y_sol = (a1*c2 - a2*c1)/det
    print(f"方程组唯一解: x={x_sol}, y={y_sol}")

    # x 数组
    x = np.linspace(x_range[0], x_range[1], points)

    # y 数组
    y1 = (c1 - a1*x)/b1
    y2 = (c2 - a2*x)/b2

    # 创建 Plotly 图表
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name=f'{a1}x + {b1}y = {c1}'))
    fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name=f'{a2}x + {b2}y = {c2}'))
    fig.add_trace(go.Scatter(x=[x_sol], y=[y_sol], mode='markers', 
                             marker=dict(color='red', size=10), name='交点'))

    fig.update_layout(title='方程组可视化（Plotly）',
                      xaxis_title='x',
                      yaxis_title='y',
                      showlegend=True)
    fig.show()


plot_linear_system_plotly((4,2,10), (3,-1,5))
```

## 线性不等式(linear inequality)

> 线性不等式 = 把等号换成不等号的一元一次方程

例子: $ 2x + 3 > 7 $

你要做的事情只有一件 : **把$x$单独留下**

---

**计算规则(和方程几乎一样)**
$$
    -3x + 6 \ge 0
$$

1. 你不是在计算一个数,而是回答一个问题
    - 哪些x,会让 $-3x + 6 \ge 0$?
2. 把常数和变量分开
    - 你需要做的第一件事不是计算而是重新组织信息
    - 我们所需要的结果 = $(关于x的部分) \ge (纯数字)$
    - 所以把6移走(两边同时减去6)
        - $ -3x + 6 \ge 0$
    - 因为两边同时减去同一个数,不会改变大小关系
3. 处理$x$前面的系数(Key)
    - 现在我们的表达式结构: $-3x \ge -6$
        - 预期得到$x在左边$
    - 所以需要同时除以-3(x前面的系数)
        - $ \frac{-3x}{-3} \ge \frac{-6}{-3}$
        - 左边变成x,右边变成2
    - 有一个物理级别的事实: 
        - 乘或除以负数,会把大小顺序反过来,所以不等号要翻转
        - $ x \le 2$

完成

--- 

唯一一条 **"新规则"**

> 不等式两边同时乘或除以一个"负数",不等号方向必须反过来

关键例子
1. $-2x > 4$
    - 如果像方程直接除以 $ x>-2$ **错误**
    - 正确做法:
        - 两边同时除以-2
        - 不等号需要反向: $x<-2$
        - 记住这个关键: $a<b \Rightarrow -a >-b$


```python
def plot_linear_inequality(a, b, c, inequality='>=', x_range=(-20, 20), num_points=500):
    """
    可视化一元线性不等式 a*x + b >= c 或 <= c
    
    参数:
    - a, b, c: 不等式参数
    - inequality: '>=' 或 '<='
    - x_range: x轴显示范围
    - num_points: 绘图精度
    """
    x = np.linspace(x_range[0], x_range[1], num_points)
    y = a*x + b - c  # 转成 y >= 0 的形式

    # 计算解区间
    if a != 0:
        if inequality == '>=':
            if a > 0:
                x_solution = np.linspace(-20, c, num_points) if a < 0 else np.linspace(c/a, x_range[1], num_points)
            else:
                x_solution = np.linspace(x_range[0], c/a, num_points)
        elif inequality == '<=':
            if a > 0:
                x_solution = np.linspace(x_range[0], c/a, num_points)
            else:
                x_solution = np.linspace(c/a, x_range[1], num_points)
    else:
        x_solution = []  # a=0 的特殊情况

    # 绘制图形
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name=f'{a}x + {b} - {c}'))

    # 绘制解区间
    if len(x_solution) > 0:
        y_fill = a*x_solution + b - c
        fig.add_trace(go.Scatter(
            x=x_solution,
            y=y_fill,
            fill='tozeroy',
            mode='none',
            fillcolor='rgba(0,200,0,0.3)',
            name='解区间'
        ))

    # 添加 y=0 参考线
    fig.add_trace(go.Scatter(
        x=[x_range[0], x_range[1]],
        y=[0,0],
        mode='lines',
        line=dict(color='black', dash='dash'),
        name='y=0'
    ))

    fig.update_layout(
        title=f'线性不等式可视化: {a}x + {b} {inequality} {c}',
        xaxis_title='x',
        yaxis_title=f'{a}x + {b} - {c}',
        showlegend=True
    )

    fig.show()

    



# 示例调用
plot_linear_inequality(a=-3, b=6, c=0, inequality='<=')
```

## 不等式组(System of inequalities)

不等式组就是多个不等式同时成立
$$
    \begin{cases}
    x - 2 \ge 0 \\
    x + 3 \le 5 
    \end{cases}
$$

1. 第一条: $x \ge 2$
2. 第二条: $x \le 2$

两条都要成立,所以最终解的是一个交集: $x=2$

核心思想 : 不等式组的解 = 每条不等式的解区间的交集

---

**计算步骤**

$$
    \begin{cases}
    2x-3 > 1\\
    -3x + 5 \ge 2
    \end{cases}
$$

1. 步骤1 : 分别解每个不等式(所有的移项都需要转换运算符)
    - $2x- 3 > 1 \Rightarrow 2x > 4 \Rightarrow x > 2$
    - $-3x + 5 \ge 2 \Rightarrow -3x \ge -3 \Rightarrow x \le 1$ (注意翻转不等号)
2. 步骤2 : 取交集
    - 第一条 : $x > 2 \Rightarrow (2,\infty)$
    - 第二条 : $x \le 1 \Rightarrow (-\infty,1)$
3. 交集 = 没有重叠 = 无解

```python
# 定义网格
x = np.linspace(0, 5, 200)
y = np.linspace(0, 5, 200)
X, Y = np.meshgrid(x, y)

# 定义不等式组
ineq1 = X + Y <= 4
ineq2 = X >= 1
ineq3 = Y >= 0

# 可行区域（所有不等式同时满足）
feasible = ineq1 & ineq2 & ineq3

# 只显示可行区域，其他区域设为 NaN
Z = np.where(feasible, 1, np.nan)

# 绘制热力图
fig = go.Figure(go.Heatmap(
    x=x,
    y=y,
    z=Z,
    colorscale='Viridis',
    showscale=False
))

fig.update_layout(
    title="不等式组可行区域示意",
    xaxis_title="x",
    yaxis_title="y",
    yaxis=dict(scaleanchor="x"),  # 保持x,y比例
)

fig.show()
```

##  绝对值方程与不等式(Absolute Value equations and inequalities)

>解集就是：所有满足方程或不等式条件的数的集合。

$$
    |x| = \begin{cases}
        x,x \ge 0\\
        -x,x<0
        \end{cases}
$$
直观理解: 绝对值表示距离,不管正负,**都取非负数**

例子: 
- $ |5| = 5 $
- $ |-3| = 3 $

---

**绝对值方程** 
$$ |x| = a $$

步骤
1. $a\ge0$,否则无解
2. 直接拆成两个方程: $x=a , x = -a$

例子
1. $|x| = 4$
    - 拆成: $x=4 , x =-4$
    - 解集: {-4,4}

绝对值方程含更多变量

$$ |2x - 3| = 5 $$
- 拆成两个方程: (通过移项)
    - $2x -3 = 5 \Rightarrow 2x = 8 \Rightarrow x =  4$
    - $2x - 3 = -5 \Rightarrow 2x = -2 \Rightarrow x = -1$
- 解集: {-1,4}

---

**绝对值不等式**

绝对值不等式通常有两类,思路不同:
1. 形式一: $|x| < a , |x| \le a$
    - $ |x| < a (a > 0) $
    - 拆成双边限制: $-a < x < a $
        - $|x| \le a \Rightarrow -a \le x \le a$
        - 理解: x的距离小于a -> 落在区间[-a,a]
    - 例子: $|x-2| < 5 $
    - 拆成: $-5 < x -2 < 5 \Rightarrow -3 < x < 7$(需要除以项的系数)
2. 形式二: $|x| > a , |x| \ge a$
    - $ |x| > a (a > 0) $
    - 拆成两边开口: $x < -a , x > a$
        - 理解: x的距离大于a -> x在区间之外
    - 例子: $|x+1| \ge 3$
    - 拆成
        1. $x+1 \le -3 \Rightarrow x \le -4$
        2. $x+1 \ge 3 \Rightarrow x \ge 2$
        3. 解集: $(-\infty ,-4] ∪ [2,\infty)$
            - 小括号 ( ) → 不包括这个端点
            - 方括号 [ ] → 包括这个端点
            - ∞ 或 -∞ → 永远用小括号，因为无穷大不是具体数字，不能包括
            - ∪ → 并集，意思是“或者”，也就是两个区间合在一起的所有 x


---

**绝对值方程/不等式技巧**
1. 方程→拆成两条等式(正负情况)
2. 不等式<或≤ → 双边限制
3. 不等式 > 或 ≥ → 两边开口，分两段
4. 检查 a 是否 ≥ 0，否则无解
